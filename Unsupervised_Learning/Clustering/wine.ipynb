```python
x=df
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,20):
    data=KMeans(n_clusters=i,init='k-means++',random_state=42)
    data.fit(x)
    wcss.append(data.inertia_)
wcss
    
```

```python
plt.bar(range(1,20),wcss)
```

```python
data1=KMeans(n_clusters=12,init='k-means++',random_state=42)
data1.fit(x)
y_pred=data1.predict(x)
y_pred
```

```python
x['Cluster']=y_pred
x
```

```python
x1=x.drop('Cluster',axis=1)
y1=x['Cluster']
x1
```

```python
y1
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model1 = RandomForestClassifier()
data2 = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'bootstrap': [True, False],
    'random_state':[None,0,1,42]
}
clf = GridSearchCV(model1, data2, cv=5, scoring='accuracy')  # Changed 'data' to 'data2'
clf.fit(x_train, y_train)

# View results
print("Best parameters:", clf.best_params_)
print("Best score:", (clf.best_score_)*100)
```

```python
model=RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=20,bootstrap=True,random_state=None)
model.fit(x_train,y_train)
y_pred1=model.predict(x_test)
y_pred1
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred1)
cm
```

```python
score=accuracy_score(y_test,y_pred1)*100
score
```

```python
report=classification_report(y_test,y_pred1)
print(report)
```
