```python
from scipy.stats import chi2_contingency
```

```python
target='Survived'
other_columns=df.select_dtypes(include='object').columns
result={}
for col in other_columns:
    p_val=chi2_contingency(pd.crosstab(df[col],df[target]))[1]
    result[col]=p_val

chi=pd.DataFrame.from_dict(result,orient='index',columns=['p_val']).sort_values(by='p_val',ascending=False)
chi
```

```python
sns.heatmap(chi,annot=True,fmt='.2f',cmap='Reds_r')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
data={'n_estimators': [100, 200, 300, 500],'max_features': ['sqrt', 'log2', None],'bootstrap': [True, False],'criterion': ['gini', 'entropy'],'random_state':[0,1,42]}
model=RandomForestClassifier()
grid=GridSearchCV(model,data,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
print('score=',grid.best_score_)
```

```python

model1=RandomForestClassifier(n_estimators=100,random_state=42,bootstrap=True,criterion='entropy',max_features=None)
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
labb=[0,1]
cmd=ConfusionMatrixDisplay(cm,display_labels=labb)
cmd.plot()
```

```python
score=accuracy_score(y_test,y_pred)*100
score
```

```python
report=classification_report(y_test,y_pred)
print(report)
```
