```python
from scipy.stats import chi2_contingency
target='Survived'
obj_columns=train_df.select_dtypes(include='object').columns
result={}
for col in obj_columns:
    p_value=chi2_contingency(pd.crosstab(train_df[col],train_df[target]))[1]
    result[col]=p_value
chi_result=pd.DataFrame.from_dict(result,orient='index',columns=['p_value']).sort_values(by='p_value',ascending=True)
chi_result
```

```python
sns.heatmap(chi_result,annot=True,cmap='Reds_r',fmt='.2e')
```
