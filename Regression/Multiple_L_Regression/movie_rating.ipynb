```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col=make_column_transformer((OneHotEncoder(handle_unknown='ignore'), ['Movie Names', 'Genere']),remainder='passthrough')
x=col.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
df1=pd.DataFrame({'actual_value':y_test,'predicted_values':y_pred,'difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)*100
score
```
