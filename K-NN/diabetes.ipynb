```python
#seperate x and y
#x=2d and y=1d
x=df.iloc[:,:-1].values #array
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#split data into training and testind data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
#Normalisation or scaling (makes values into same range)
#higher column priority kodum...ath paadilla

#methods===>

#1. Standard Scalar [z=(x-u)/s]
#2. Minmax Scalar
```

```python
#1. Standard Scalar
#z=(x-u)/s
#z=Output
# x=input
# u=mean of training data
# s=standard deviation
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Model creation
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
print(knn.predict(scalar.transform([[3,110,72,34,3,24.3,0.123,54]])))
```

```python
y_test
```

```python
#Mode Performance Evaluation
#Methods=====>
#1. Accuracy score
# 2.Recall
# 3.Precision
# 4.F1_Score
# Confusion Matrix based......for classification model
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Accuracy Score........
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
