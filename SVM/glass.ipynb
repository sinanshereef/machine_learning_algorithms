```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
neigh=KNeighborsClassifier(n_neighbors=5)
naive=GaussianNB()
sv=SVC()
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score
lst=[neigh,naive,sv]
```

```python
for i in lst:
    print('The Model is',i)
    print("*"*100)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    print('Predicted Values are',y_pred)
    print('*'*100)
    print('output test value=',y_test)
    print('*'*100)
    print('****Confusion Matrix*****')
    print('Confuson Matrix is', confusion_matrix(y_test,y_pred))
    print('*'*100)
    lab=[2,1,6,5,3,7]
    print('*****confusion matrix display*****')
    cmd=ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred),display_labels=lab)
    print(cmd.plot())
    print('*'*100)
    print('Accuracy score of the Mode =',accuracy_score(y_test,y_pred)*100)
    
```
