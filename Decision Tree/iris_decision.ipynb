```python
df.drop('Id',axis=1,inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
lab=['Iris-virginica', 'Iris-setosa','Iris-versicolor']
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```

```python
score=accuracy_score(y_test,y_pred)*100
score
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model,feature_names=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"],class_names=['Iris-virginica', 'Iris-setosa','Iris-versicolor'],filled=True)
```
